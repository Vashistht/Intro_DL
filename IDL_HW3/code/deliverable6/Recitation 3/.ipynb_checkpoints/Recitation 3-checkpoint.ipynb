```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

Load a sample image of a hog/pig as our benign input (i.e. input without the presence of adversaries). Inputs to the pre-trained model is 224 by 224, so we need to resize them. Note that inputs to ML models may have dimensions defined differently from human readable images, and you may need to refer to the spec of ML models.

```python
from PIL import Image
from torchvision import transforms

# read the image, resize to 224 and convert to PyTorch Tensor
pig_img = Image.open("pig.jpg")
preprocess = transforms.Compose([
   transforms.Resize(224),
   transforms.ToTensor(),
])
pig_tensor = preprocess(pig_img)[None,:,:,:]

# plot image (note that numpy using HWC whereas Pytorch user CHW, so we need to convert)
plt.imshow(pig_tensor[0].numpy().transpose(1,2,0))
plt.xticks([])
plt.yticks([])
plt.show()
```

Looks good. Now let's load the pre-trained ResNet50 model and apply it to the image, after necessary transforms (the weird indexing here is just used to comply with PyTorch standards that all inputs to modules should be of the form `batch_size x num_channels x height x width`).

Image normalization is a common preprocessing step. It ensures consistent data scaling.

Question: why we need to Normalize as a Module? Why not using existing packages?

```python
import torch
import torch.nn as nn
from torchvision.models import resnet50

# simple Module to normalize an image
class Normalize(nn.Module):
    def __init__(self, mean, std):
        super(Normalize, self).__init__()
        self.mean = torch.Tensor(mean)
        self.std = torch.Tensor(std)
    def forward(self, x):
        return (x - self.mean.type_as(x)[None,:,None,None]) / self.std.type_as(x)[None,:,None,None]

# values are standard normalization for ImageNet images, 
# from https://github.com/pytorch/examples/blob/master/imagenet/main.py
norm = Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])

# load pre-trained ResNet50, and put into evaluation mode (necessary to e.g. turn off batchnorm)
model = resnet50(pretrained=True)
model.eval();
```

```python
# form predictions
pred = model(norm(pig_tensor))
print (pred.shape)
```

Pred now contains a 1000 dimensional vector containing the class logits for the 1000 imagenet classes (i.e., if you wanted to convert this to a probability vector, you would apply the softmax operator to this vector).  To find the highest likelihood class, we simply take the index of maximum value in this vector, and we can look this up in a list of imagenet classes to find the corresponding label.

```python
import json
with open("imagenet_class_index.json") as f:
    imagenet_classes = {int(i):x[1] for i,x in json.load(f).items()}
print(imagenet_classes[pred.max(dim=1)[1].item()])
```

Loooking good! (note that ImageNet has one class correpsonding to both "hog" and "pig", so this is the correct label).  We should note that this is the first pig image we tried here, so it doesn't take any tweaking to get a result like this ... modern image classifiers are pretty impressive.

Fast Gradient Sign Method (FGSM, https://arxiv.org/abs/1412.6572): 
$$
x'=clip(x-sign(\frac{\partial (-L(x,y))}{\partial x})*\epsilon)
$$

$x $: original benign input

$y $: ground truth label

$L $: loss function respect to inputs and a class (e.g. Cross Entropy)

$clip$: clip function with respect to $x$

$sign$: the sign function, returns -1, 0, or 1

$\epsilon$: the $L^{\infty}$ distance

$x'$ the adversarial example generated

```python
epsilon = 36./255

adv_tensor=pig_tensor.detach().clone()
adv_tensor.requires_grad = True#we want gradients for this input
pred = model(norm(adv_tensor))#normalized inputs
model.zero_grad()
loss=-nn.CrossEntropyLoss()(pred,torch.LongTensor([341]))#341 is the hog/pig class; notice the sign here.
loss.backward()#we only compute the gradient; we do not update the models
grads=adv_tensor.grad
with torch.no_grad():
    adv_tensor = adv_tensor - epsilon*grads.sign()# Q: why we only take the sign? Hint: which Lp is used?
    eta = torch.clamp(adv_tensor - pig_tensor, min=-epsilon, max=epsilon)#ensure the input is within L_inf distance
    adv_tensor = torch.clamp(pig_tensor + eta, min=0, max=1).detach().clone()#ensure the input is a valid image
    adv_tensor = torch.round(adv_tensor*255)/255#ensure the input is in 8-bit RGB format
pred = model(norm(adv_tensor))
print("True class probability:", nn.Softmax(dim=1)(pred)[0,341].item())
```

Now the image is no longer a pig/hog

```python
max_class = pred.max(dim=1)[1].item()
print ("Predicted class index: ", max_class)
print("Predicted class: ", imagenet_classes[max_class])
print("Predicted probability:", nn.Softmax(dim=1)(pred)[0,max_class].item())
```

Fast Gradient Sign Method (targeted): 
$$
x'=clip(x-sign(\frac{\partial (L(x,t))}{\partial x})*\epsilon)
$$

$t $: target class (e.g. West_Highland_white_terrier)

we plot the probability that t is predicted with respect to different choices of $\epsilon$

```python
probabilities=list()
for eps in range(4,40):#plot epsilon curves
    epsilon=eps/255.

    adv_tensor=pig_tensor.detach().clone()
    adv_tensor.requires_grad = True
    pred = model(norm(adv_tensor))
    model.zero_grad()
    loss=nn.CrossEntropyLoss()(pred,torch.LongTensor([203]))# 203 is West_Highland_white_terrier
    loss.backward()
    grads=adv_tensor.grad
    with torch.no_grad():
        adv_tensor = adv_tensor - epsilon*grads.sign()
        eta = torch.clamp(adv_tensor - pig_tensor, min=-epsilon, max=epsilon)
        adv_tensor = torch.clamp(pig_tensor + eta, min=0, max=1).detach().clone()
        adv_tensor = torch.round(adv_tensor*255)/255
    pred = model(norm(adv_tensor))
    probabilities.append(nn.Softmax(dim=1)(pred)[0,203].item())#target class probability
plt.plot(np.arange(4,40),probabilities,label="FGSM")
plt.ylim([0,1])
plt.ylabel("Target class probability")
plt.xlabel("eps (/255.)")
plt.legend()
plt.show()
```

Iterative Fast Gradient Sign Method (I-FGSM, https://arxiv.org/abs/1607.02533): 
$$
x'_i=clip(x'_{i-1}-sign(\frac{\partial (L(x'_{i-1},t))}{\partial x'_{i-1}})*\epsilon)
$$

```python
probabilities2=list()
for eps in range(4,40):
    epsilon=eps/255.

    adv_tensor=pig_tensor.detach().clone()
    for i in range(5):
        adv_tensor.requires_grad = True
        pred = model(norm(adv_tensor))
        model.zero_grad()
        loss=nn.CrossEntropyLoss()(pred,torch.LongTensor([203]))
        loss.backward()
        grads=adv_tensor.grad
        with torch.no_grad():
            adv_tensor = adv_tensor - epsilon*grads.sign()
            eta = torch.clamp(adv_tensor - pig_tensor, min=-epsilon, max=epsilon)
            adv_tensor = torch.clamp(pig_tensor + eta, min=0, max=1).detach().clone()
            adv_tensor = torch.round(adv_tensor*255)/255.
    pred = model(norm(adv_tensor))
    probabilities2.append(nn.Softmax(dim=1)(pred)[0,203].item())
plt.plot(np.arange(4,40),probabilities,label="FGSM")
plt.plot(np.arange(4,40),probabilities2,label="I-FGSM")
plt.ylabel("Target class probability")
plt.xlabel("eps (/255.)")
plt.legend()
plt.show()
```

Now the attacks is much stronger, but there is still too much fluctuation.

What is one disadvantage to perturb inputs always by the maximum $L_{\infty}$ distance $\epsilon$?













Projected Gradient Descent (PGD, https://arxiv.org/abs/1706.06083): 
$$
x'_i=clip(x'_{i-1}-sign(\frac{\partial (L(x'_{i-1},t))}{\partial x'_{i-1}})*\alpha)
$$

$\alpha$: step size of perturbation, analogous to learning rate.

Question:
1. What is the disadvantage of a too large learning rate?
2. What is the disadvantage of a too small learning rate?


```python
probabilities3=list()
alpha=8./255
for eps in range(4,40):
    epsilon=eps/255.

    adv_tensor=pig_tensor.detach().clone()
    for i in range(5):
        adv_tensor.requires_grad = True
        pred = model(norm(adv_tensor))
        model.zero_grad()
        loss=nn.CrossEntropyLoss()(pred,torch.LongTensor([203]))
        loss.backward()
        grads=adv_tensor.grad
        with torch.no_grad():
            adv_tensor = adv_tensor - alpha*grads.sign()
            eta = torch.clamp(adv_tensor - pig_tensor, min=-epsilon, max=epsilon)
            adv_tensor = torch.clamp(pig_tensor + eta, min=0, max=1).detach().clone()
            adv_tensor = torch.round(adv_tensor*255)/255.
    pred = model(norm(adv_tensor))
    probabilities3.append(nn.Softmax(dim=1)(pred)[0,203].item())
plt.plot(np.arange(4,40),probabilities,label="FGSM")
plt.plot(np.arange(4,40),probabilities2,label="I-FGSM")
plt.plot(np.arange(4,40),probabilities3,label="PGD")
plt.ylabel("Target class probability")
plt.xlabel("eps (/255.)")
plt.legend()
plt.show()
```

Now the curve is smoother, but the attack is not good enough when $\epsilon$ is small (i.e. $\epsilon<=\alpha$). How might we improve the attack besides choosing different values of $\alpha$? What does the Adam optimizer do (https://arxiv.org/abs/1412.6980)?

```python
probabilities4=list()
for eps in range(4,40):
    epsilon=eps/255.

    adv_tensor=pig_tensor.detach().clone()
    m=torch.zeros(pig_tensor.shape)
    v=torch.zeros(pig_tensor.shape)
    
    for i in range(5):
        adv_tensor.requires_grad = True
        pred = model(norm(adv_tensor))
        model.zero_grad()
        loss=nn.CrossEntropyLoss()(pred,torch.LongTensor([203]))
        loss.backward()
        grads=adv_tensor.grad
        with torch.no_grad():
            #a manually coded Adam optimizer
            t=i+1
            m=0.9*m+0.1*grads
            v=0.999*v+0.001*grads*grads
            mhat=m/(1.0 - 0.9**t)
            vhat=v/(1.0 - 0.999**t)
            grads=mhat / (torch.sqrt(vhat) + 1e-8)
            adv_tensor = adv_tensor - 8./255*grads.sign()
            eta = torch.clamp(adv_tensor - pig_tensor, min=-epsilon, max=epsilon)
            adv_tensor = torch.clamp(pig_tensor + eta, min=0, max=1).detach().clone()
            adv_tensor = torch.round(adv_tensor*255)/255.
    pred = model(norm(adv_tensor))
    probabilities4.append(nn.Softmax(dim=1)(pred)[0,203].item())
plt.plot(np.arange(4,40),probabilities,label="FGSM")
plt.plot(np.arange(4,40),probabilities2,label="I-FGSM")
plt.plot(np.arange(4,40),probabilities3,label="PGD")
plt.plot(np.arange(4,40),probabilities4,label="PGD, with Adam")
plt.ylabel("Target class probability")
plt.xlabel("eps (/255.)")
plt.legend()
plt.show()
```

What does our adversarial example look like now?

```python
epsilon=4/255.
adv_tensor=pig_tensor.detach().clone()
m=torch.zeros(pig_tensor.shape)
v=torch.zeros(pig_tensor.shape)
    
for i in range(5):
    adv_tensor.requires_grad = True
    pred = model(norm(adv_tensor))
    model.zero_grad()
    loss=nn.CrossEntropyLoss()(pred,torch.LongTensor([203]))
    loss.backward()
    grads=adv_tensor.grad
    with torch.no_grad():
        t=i+1
        m=0.9*m+0.1*grads
        v=0.999*v+0.001*grads*grads
        mhat=m/(1.0 - 0.9**t)
        vhat=v/(1.0 - 0.999**t)
        grads=mhat / (torch.sqrt(vhat) + 1e-8)
        adv_tensor = adv_tensor - 8./255*grads.sign()
        eta = torch.clamp(adv_tensor - pig_tensor, min=-epsilon, max=epsilon)
        adv_tensor = torch.clamp(pig_tensor + eta, min=0, max=1).detach().clone()
        adv_tensor = torch.round(adv_tensor*255)/255.
plt.imshow((adv_tensor)[0].detach().numpy().transpose(1,2,0))
plt.xticks([])
plt.yticks([])
plt.show()
```
